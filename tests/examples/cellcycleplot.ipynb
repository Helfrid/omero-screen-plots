```python



import pandas as pd

from omero_screen_plots.cellcycleplot import cellcycle_plot

df = pd.read_csv("../test_data.csv")
conditions = ["ctr", "palb"]
```

```python
for cell_line in df.cell_line.unique():
    cellcycle_plot(
        df=df,
        conditions=conditions,
        condition_col="condition",  # default
        selector_col="cell_line",  # default
        selector_val=cell_line,
        title=f"{cell_line} cell cycle plots",
        save=False,
        path=None,
)
```

```python
# check for 2 and 1 plate

df1 = df[df.plate_id == 1969]
df2 = df[df.plate_id.isin([1947, 1969])]

cellcycle_plot(
    df=df1,
    conditions=conditions,
    condition_col="condition",  # default
    selector_col="cell_line",  # default
    selector_val="RPE1wt",
    title="RPE1wt cell cycle plots",
    save=False,
    path=None,
)

cellcycle_plot(
    df=df2,
    conditions=conditions,
    condition_col="condition",  # default
    selector_col="cell_line",  # default
    selector_val="RPE1wt",
    title="RPE1wt cell cycle plots",
    save=False,
    path=None,
)
```
