```python
import pandas as pd

from omero_screen_plots.cellcycleplot import stacked_barplot

df = pd.read_csv("../test_data.csv")
conditions = ["ctr", "palb"]
```

```python
stacked_barplot(
    df,
    condition_col="condition",
    conditions=conditions,
    selector_col="cell_line",
    selector_val="RPE1wt",
    title="RPE1 WT stacked barplot",
    save=False,
    path=None,
)
```
